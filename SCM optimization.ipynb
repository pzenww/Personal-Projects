```python
import gurobipy as gp
from gurobipy import GRB
```

```python
try:
    m = gp.Model('practice')

    farm = [1,2]
    elev = [3,4,5]
    mill = [6,7,8]

    for i in farm:
        for j in elev:
            globals()[f'x{i}{j}'] = m.addVar(vtype=GRB.CONTINUOUS, name=f'x{i}{j}')
    for i in elev:
        for j in mill:
            globals()[f'x{i}{j}'] = m.addVar(vtype=GRB.CONTINUOUS, name=f'x{i}{j}')

    farm_supply = [300,300]
    mill_demand = [200,100,300]

    for i in farm:
        m.addConstr(gp.quicksum(globals()[f'x{i}{j}'] for j in elev) == farm_supply[i-1], name=f'c{j}')
    for j in mill:
        m.addConstr(gp.quicksum(globals()[f'x{i}{j}'] for i in elev) == mill_demand[j-6], name=f'c{j-3}')

    
    for j in elev:
        m.addConstr(gp.quicksum(globals()[f'x{i}{j}'] for i in farm) == gp.quicksum(globals()[f'x{j}{k}'] for k in mill), name=f'c{j+3}')

    fe_cost = [[16,10,12],[15,14,17]]
    em_cost = [[6,8,10],[7,11,11],[4,5,12]]

    obj = (gp.quicksum(fe_cost[i-1][j-3] * globals()[f'x{i}{j}'] for i in farm for j in elev) 
           + gp.quicksum(em_cost[i-3][j-6] * globals()[f'x{i}{j}'] for i in elev for j in mill))

    m.setObjective(obj, GRB.MINIMIZE)

    m.optimize()
    
    for v in m.getVars():
        print(f"{v.VarName} {v.X:g}")

    print(f"Obj: {m.ObjVal:g}")

except gp.GurobiError as e:
    print(f"Error code {e.errno}: {e}")

except AttributeError:
    print("Encountered an attribute error")
```
